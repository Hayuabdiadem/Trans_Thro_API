```python
import pandas as pd
import time

from datetime import datetime
```

```python
df_trans = pd.read_csv("raw_transactions - consolidated.csv")
```

```python
df_trans
```

```python
#df_trans['trans_date'] = datetime.fromtimestamp(df_trans['timestamp']).apply(lambda x: x.to_datetime64())
df_trans['trans_datetime'] = pd.to_datetime(df_trans['timestamp'], unit='s') #.apply(lambda x: x.to_datetime64())
df_trans
```

```python
df_market_min = pd.read_csv("MarketData_2017_2021.csv") #BTCUSD_minute.csv")
```

```python
df_market_min
```

```python
df_market_min['market_datetime'] = pd.to_datetime(df_market_min['Timestamp'], unit='s') #.apply(lambda x: x.to_datetime64())
```

```python
#Create date and hour columns on both datasets

df_trans['trans_date'] = df_trans['trans_datetime'].dt.date
df_trans['trans_hour'] = df_trans['trans_datetime'].dt.hour
#df_trans

df_market_min['trans_date'] = df_market_min['market_datetime'].dt.date
df_market_min['trans_hour'] = df_market_min['market_datetime'].dt.hour
df_market_min
```

```python
#aggregate both the sources to hour level

df_trans_hour = df_trans.groupby (by = ['address', 'hash_id', 'from_owner_type', 'to_owner_type', 'trans_date', 'trans_hour'], as_index=False).sum()
df_trans_hour.drop(['sno', 'timestamp'], axis=1, inplace=True)
df_trans_hour
df_market_hour = df_market_min.groupby (by = ['trans_date', 'trans_hour'], as_index=False).mean()
df_market_hour.drop(['Timestamp'], axis=1, inplace=True)
df_market_hour
```

```python
#test
#df_trans_hour[df_trans_hour['hash_id'] == '00019b61eb77024c59c098f14d57cf373c53053cc652903ebe5dfeaefbfee700']

df_market_hour[df_market_hour['trans_date'] == pd.Timestamp(2021,3,22)]

#min(df_market_hour['trans_date'])
```

```python
#Join both the tables
df_trans_market = pd.merge(df_trans_hour, df_market_hour, on =['trans_date', 'trans_hour'], how = 'left')
df_trans_market
```
